```python
import numpy as np
import matplotlib.pyplot as plt
import math 
```

```python
print (math.log2(1))
```

```python

x = np.linspace(1, 1000, 1000)
plt.title("Linear Growth Rates")
plt.xlabel("N")
plt.ylabel("Time")

plt.plot(x, x,label="n")
plt.plot(x,x/2,label="n / 2")
plt.legend();
plt.show()
plt.clf()
```

```python
plt.title("O N Squared Growth Rates")
x = np.linspace(1, 1000, 10000)
plt.xlabel("N")
plt.ylabel("Time")

def f(x):
	y = []
	for i in x:
		y.append(i*(math.log2(i)))
	return y 

y = f(x)
plt.plot(x, x*x, label="n^2")
plt.plot(x, (x*x)/2, label=("N^2 / 2"))
plt.legend()
plt.show()
plt.clf()
```

```python
plt.title("O log(N) Growth Rates")
x = np.linspace(1, 1000000,100)
plt.xlabel("N")
plt.ylabel("Time")


def f(x):
	y = []
	for i in x:
		y.append((math.log(i)))
	return y


def f2(x):
	y = []
	for i in x:
		y.append((math.log2(i)))
	return y


def f10(x):
	y = []
	for i in x:
		y.append((math.log10(i)))
	return y
plt.plot(x, f2(x), label="log base 2")
plt.plot(x, f(x), label="natural log")
plt.plot(x, f10(x), label="log base 10")
y = f(x)
plt.legend()
plt.show()
plt.clf()
```
